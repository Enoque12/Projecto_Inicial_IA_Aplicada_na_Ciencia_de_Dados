## Aula 02 - Visualizando os dados

### Mão na Massa - Gerando visuais em relação a localidade

#### Desafio 1:

Construa e execute um prompt no ChatGPT para gerar a partir de nossa base de dados um visual que apresente o faturamento das vendas por região. Lembre-se que aqui o faturamento é dado pelo produto do valor unitário pela quantidade comprada somado ao frete.

> #### <strong><font color="#203f75" size= 6>Prompt:</font></strong>

Temos um DataFrame chamado "df_final" e queremos construir um visual de gráfico de barras horizontais que apresente o faturamento das vendas por região da.

Em "df_final", temos as colunas:
* regiao: região
* preco_unitario: preço de uma unidade do produto
* quantidade: quantidade vendida do produto
* frete: frete a ser pago no total da compra.

Desejamos criar uma visualização que apresente o faturamento de cada região, em que o faturamento é dado pelo produto entre o preço unitário e a quantidade somadas ao frete.

Utilize as bibliotecas de visualização de dados do Python para criar essa visualização.

```python
# Calculando o faturamento por linha
df_final['faturamento'] = (df_final['preco_unitario'] * df_final['quantidade']) + df_final['frete']
```

```python
# Agrupando por região e somando o faturamento
faturamento_por_regiao = df_final.groupby('regiao')['faturamento'].sum().sort_values()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Estilo do gráfico
sns.set(style="whitegrid")
```

```python
# Criando o gráfico de barras horizontais
plt.figure(figsize=(8, 5))
sns.barplot(x=faturamento_por_regiao.values, y=faturamento_por_regiao.index, palette="viridis")

# Títulos e rótulos
plt.title('Faturamento por Região', fontsize=14, fontweight='bold')
plt.xlabel('Faturamento (R$)', fontweight='bold')
plt.ylabel('Região', fontweight='bold')

# Exibir valores no gráfico
for i, valor in enumerate(faturamento_por_regiao.values):
    plt.text(valor, i, f'R${valor:,.2f}', va='center')

sns.despine()
plt.grid(False)
plt.tight_layout()
plt.show()
```

#### Desafio 2:

A Zoop quer realizar uma campanha nos estados de São Paulo e Rio de Janeiro, para isso, precisa saber o perfil de seu público de acordo com o sexo biológico. Construa e execute um prompt no ChatGPT para gerar a partir de nossa base de dados um visual que apresente a composição do público por sexo biológico nos estados de São Paulo e Rio de Janeiro. Lembre-se de pedir para agrupar ou empilhar os dados para apresentação.

> #### <strong><font color="#203f75" size= 6>Prompt:</font></strong>

Temos um DataFrame chamado "df_final" e queremos construir um visual de gráfico de barras empilhadas que apresente a composição do público por sexo biológico nos estados de São Paulo e Rio de Janeiro.

Em "df_final", temos as colunas:

- sexo_biologico: sexo biológico do cliente
- uf: Estados

Desejamos criar uma visualização que apresente a composição do público por sexo biológico, agrupando pelos estados correspondente.

Utilize as bibliotecas de visualização de dados do Python para criar essa visualização.

```python
# 1. Filtrar apenas SP e RJ
df_filtrado = df_final[df_final['uf'].isin(['São Paulo', 'Rio de Janeiro'])]
```

```python
# 2. Contar clientes por sexo e estado
contagem = df_filtrado.groupby(['uf', 'sexo_biologico']).size().unstack().fillna(0)
```

```python
contagem
```

```python
# 3. Plotar gráfico de barras empilhadas
contagem.plot(kind='bar', stacked=True, figsize=(8, 6), colormap='tab10')

# 4. Personalização do gráfico
plt.title('Composição do Público por Sexo Biológico - SP e RJ')
plt.xlabel('Estado')
plt.ylabel('Número de Clientes')
plt.xticks(rotation=0)
plt.legend(title='Sexo Biológico')
plt.tight_layout()

# 5. Exibir gráfico
plt.show()
```

#### Desafio 1:

A Zoop está em processo de lançamento de sua carteira digital, que dará acesso a uma série de benefícios para sua clientela, e gostaria de saber qual a porcentagem de pessoas que já possuem pré-cadastro para este produto. O e-commerce busca compreender esse número para definir a necessidade de divulgação e esforço para chamar mais clientes para esta funcionalidade.

Construa e execute um prompt no ChatGPT para gerar a partir de nossa base de dados um visual que apresente a proporção de pessoas pré-cadastradas na Zoop Pay.

> #### <strong><font color="#203f75" size= 6>Prompt:</font></strong>

Temos um DataFrame chamado "df" e queremos construir um visual que apresente a proporção dos(as) usuários(as) que estão pre-cadastrados ao nosso serviço de Zoop_pay para finalidades de receber benefícios.

Em "df_final", temos a coluna "zoop_pay" que dispõe de tais dados.

Utilize as bibliotecas de visualização de dados do Python para criar essa visualização.

```python
# Contagem de usuários por status do zoop_pay
contagem = df_final['zoop_pay'].value_counts()

# Criar gráfico de pizza
plt.figure(figsize=(6, 6))
plt.pie(
    contagem,
    labels=contagem.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#66c2a5', '#fc8d62']  # verde e laranja, por exemplo
)

plt.title('Proporção de Usuários Pré-Cadastrados no Zoop_pay')
plt.axis('equal')  # Garante formato circular
plt.show()
```

Temos um DataFrame chamado "df_final" e queremos construir um visual que apresente a distribuição das idades dos clientes por região

Em "df_final", temos a coluna "idade" com as idades dos clientes e a coluna "regiao" com as respectivas regiões dos clientes.

Desejamos criar uma visualização que apresente a distribuição desses dados.
