## Aula 02 - Visualizando os dados

### Mão na Massa - Gerando visuais em relação a localidade

#### Desafio 1:

Construa e execute um prompt no ChatGPT para gerar a partir de nossa base de dados um visual que apresente o faturamento das vendas por região. Lembre-se que aqui o faturamento é dado pelo produto do valor unitário pela quantidade comprada somado ao frete.

> #### <strong><font color="#203f75" size= 6>Prompt:</font></strong>

Temos um DataFrame chamado "df_final" e queremos construir um visual de gráfico de barras horizontais que apresente o faturamento das vendas por região da.

Em "df_final", temos as colunas:
* regiao: região
* preco_unitario: preço de uma unidade do produto
* quantidade: quantidade vendida do produto
* frete: frete a ser pago no total da compra.

Desejamos criar uma visualização que apresente o faturamento de cada região, em que o faturamento é dado pelo produto entre o preço unitário e a quantidade somadas ao frete.

Utilize as bibliotecas de visualização de dados do Python para criar essa visualização.

````python
# Calculando o faturamento por linha
df_final['faturamento'] = (df_final['preco_unitario'] * df_final['quantidade']) + df_final['frete']
````

````python
# Agrupando por região e somando o faturamento
faturamento_por_regiao = df_final.groupby('regiao')['faturamento'].sum().sort_values()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Estilo do gráfico
sns.set(style="whitegrid")
````

````python
# Criando o gráfico de barras horizontais
plt.figure(figsize=(8, 5))
sns.barplot(x=faturamento_por_regiao.values, y=faturamento_por_regiao.index, palette="viridis")

# Títulos e rótulos
plt.title('Faturamento por Região', fontsize=14, fontweight='bold')
plt.xlabel('Faturamento (R$)', fontweight='bold')
plt.ylabel('Região', fontweight='bold')

# Exibir valores no gráfico
for i, valor in enumerate(faturamento_por_regiao.values):
    plt.text(valor, i, f'R${valor:,.2f}', va='center')

sns.despine()
plt.grid(False)
plt.tight_layout()
plt.show()
````

> #### <strong><font color="#203f75" size= 6>Prompt:</font></strong>

Agora, vamos refinar o gráfico de rosca que produzimos para a proporção de pessoas pré-cadastradas na Zoop Pay.

O nosso gráfico precisa seguir a identidade visual da Zoop, com o fundo da figura e do plot na cor CINZA_1, a cor da fatia de valor "Não" igual a CINZA_3, a cor da fatia de valor "Sim" igual a AZUL_1 e o tamanho da figura deve ser de (6, 6).

O título da figura é "Participam do programa de cashback?" com tamanho 18 e sua cor CINZA_5.

Os rótulos das fatias precisam ter a mesma cor das fatias e tamanho 12 e as porcentagens precisam estar na cor BRANCO, também com tamanho 12.

A propriedade da largura da rosca (wedgeprops) precisa ser de 0.4, a posição das porcentagens dentro da fatia (pctdistance) precisa ser de 0.8, o ângulo de início de 180 (startangle) e sem explosão de fatias.

Utilize as bibliotecas de visualização de dados do Python para criar essa visualização.

````python
# Contagem dos valores "Sim" e "Não" no Zoop Pay
contagemPay = df_final['zoop_pay'].value_counts()

# Mapeando cores
cores_fatias = [CINZA_3 if label == "Não" else AZUL_1 for label in contagem.index]

# Criando figura
fig, ax = plt.subplots(figsize=(6, 6))
fig.patch.set_facecolor(CINZA_1)  # Fundo da figura
ax.set_facecolor(CINZA_1)         # Fundo do plot

# Gráfico de rosca
wedges, texts, autotexts = ax.pie(
    contagem,
    labels=contagemPay.index,
    colors=cores_fatias,
    autopct='%1.1f%%',
    startangle=180,
    wedgeprops={'width': 0.4},
    pctdistance=0.8
)

# Ajustando cor e tamanho dos rótulos (labels)
for text, color in zip(texts, cores_fatias):
    text.set_color(color)
    text.set_fontsize(12)

# Ajustando cor e tamanho das porcentagens
for autotext in autotexts:
    autotext.set_color(BRANCO)
    autotext.set_fontsize(12)

# Título
ax.set_title("Proporção de Usuários Pré-Cadastrados no Zoop_pay", fontsize=18, color=CINZA_5)

plt.tight_layout()
plt.show()
````
